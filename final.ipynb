```python
cat_vars=['BusinessTravel', 'Department', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime']
for var in cat_vars: cat_list='var'+'_'+ var
cat_list=pd.get_dummies(emp[var], prefix=var)
emp1=emp.join(cat_list)
emp=emp1
```

```python
cat_vars=['BusinessTravel', 'Department', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime']
emp_vars=emp.columns.values.tolist()
to_keep=[i for i in emp_vars if i not in cat_vars]
```

```python
emp_final=emp[to_keep]
emp_final.columns.values
```

```python
emp_final_vars=emp_final.columns.values.tolist()
Y=['Attrition']
X=[i for i in emp_final_vars if i not in Y]
```

```python
from sklearn import datasets 
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression 

model=LogisticRegression()

rfe = RFE(model, 15)
rfe = rfe.fit(emp_final[X], emp_final[Y])
print(rfe.support_)
print(rfe.ranking_)
```
